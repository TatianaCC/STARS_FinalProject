EJEMPLO: Birch Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pickle import dump, load
from sklearn.datasets import make_blobs
from sklearn.cluster import Birch


X_train = pd.read_csv('../Samples/Clean/Feature_Selection/X_train.csv')
```

```python
#Convertir el DataFrame completo a un array NumPy
X_train_array = X_train.to_numpy()

#Crear y ajustar el modelo Birch - ajustar los parámetros threshold y branching_factor según necesidades
model = Birch(threshold=0.5, branching_factor=50, n_clusters=None)
model.fit(X_train_array)

# Obtener los labels de los clústeres
labels = model.labels_

# Visualizar los resultados (seleccionando dos características para plotear)
plt.figure(figsize=(12, 8))
plt.scatter(X_train_array[:, 0], X_train_array[:, 1], c=labels, cmap='rainbow', marker='o', s=1)
plt.title('Birch Clustering')
plt.xlabel('Glon')
plt.ylabel('Glat')
plt.show()
```

```python
dump(model, open("../Models/birch.pkl", "wb"))
```

"n_clusters": Si se establece en None, Birch no realiza un clustering final global. Si se especifica un número entero, Birch realizará un clustering final global con el número dado de clústeres."
Ahora vamos a probar estableciendo el número de clústers en 87, que es la cantidad que tenemos. 

```python
#Convertir el DataFrame completo a un array NumPy
X_train_array = X_train.to_numpy()

#Crear y ajustar el modelo Birch
# Puedes ajustar los parámetros threshold y branching_factor según tus necesidades
model = Birch(threshold=0.5, branching_factor=50, n_clusters=87)
model.fit(X_train_array)

# Obtener los labels de los clústeres
labels = model.labels_

# Visualizar los resultados (selecciona dos características para plotear)
plt.scatter(X_train_array[:, 0], X_train_array[:, 1], c=labels, cmap='rainbow', marker='o', s=1)
plt.title('Birch Clustering')
plt.xlabel('Glon')
plt.ylabel('Glat')
plt.show()
```
