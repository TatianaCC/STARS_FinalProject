```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pickle import dump, load
from sklearn.datasets import make_blobs
from sklearn.cluster import OPTICS


X_train = pd.read_csv('../Samples/Clean/Feature_Selection/X_train.csv')
X_test = pd.read_csv('../Samples/Clean/Feature_Selection/X_test.csv')
y_train = pd.read_csv('../Samples/Clean/Feature_Selection/Y_train.csv')
y_test = pd.read_csv('../Samples/Clean/Feature_Selection/Y_test.csv')
```

```python
# Ver los nombres de las columnas
print("Nombres de las columnas:")
print(X_train.columns)
```

```python
# Convertir el DataFrame completo a un array NumPy
X_train_array = X_train.to_numpy()

# Crear y ajustar el modelo OPTICS
model = OPTICS(min_samples=5, xi=0.05, min_cluster_size=0.1)
model.fit(X_train_array)

# Obtener los labels de los clústeres
labels = model.labels_

#Visualizar los resultados (selecciona dos características para plotear)
plt.scatter(X_train_array[:, 0], X_train_array[:, 1], c=labels, cmap='rainbow', marker='o', s=1)
plt.title('OPTICS Clustering')
plt.xlabel('Glon')
plt.ylabel('Glat')
plt.show()
```

OPTICS (Ordering Points To Identify the Clustering Structure) no tiene un método predict porque es un algoritmo de clustering basado en densidad similar a DBSCAN. En lugar de predecir etiquetas para nuevos datos, OPTICS se utiliza principalmente para identificar la estructura de densidad de un conjunto de datos.
